```python
import os
from dotenv import load_dotenv

load_dotenv()


GROQ_API_KEY = os.getenv('GROQ_API_KEY')
```

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader
```

```python
loader_pdf = PyPDFDirectoryLoader(
    "../KnowledgeBases"
)

docs_pdf = loader_pdf.load()
len(docs_pdf)
```

### Split the docuements

```python
pip install langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
texts = text_splitter.split_documents(docs_pdf)

len(texts)
```

### Embeddings

```python
pip install langchain-huggingface sentence-transformers tf-keras
```

```python
from langchain_community.embeddings import HuggingFaceBgeEmbeddings

model_name = "BAAI/bge-base-en"
encode_kwargs = {'normalize_embeddings': True}

bge_embeddings = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs={'device': 'cuda'},
    encode_kwargs=encode_kwargs
)
```

### Store in chroma VectorDB 

```python
pip install langchain_chroma
```

```python
from langchain_chroma import Chroma

persist_directory = 'kb_db'

## Here you can change the embeddings etc
embedding = bge_embeddings

vectordb = Chroma.from_documents(documents=texts,
                                 embedding=embedding,
                                 persist_directory=persist_directory)
```

### Testing the RAG

```python
pip install langchain-groq
```

```python
retriever = vectordb.as_retriever(search_kwargs={"k": 5})
```

```python
from langchain_core.prompts import ChatPromptTemplate, PromptTemplate
from langchain_core.runnables import RunnablePassthrough

from langchain_core.output_parsers import StrOutputParser
from langchain_core.output_parsers import JsonOutputParser
```

```python
from langchain_groq import ChatGroq

GROQ_LLM = ChatGroq(
            model="qwen/qwen3-32b",
        )
```

```python
rag_context_agent_prompt = PromptTemplate(
    template="""<|begin_of_text|><|start_header_id|>system<|end_header_id|>
You are the Context Retrieval Agent for SafeLora. 
Your job is ONLY to extract and organize relevant information from the retrieved CONTEXT. 
You do NOT generate advice, diagnosis, recommendations, or IPM steps. 
You only filter, extract, and structure raw factual information.

Your task:
1. Read all retrieved CONTEXT.
2. Extract only the information that relates to the given:
   - plant,
   - disease,
   - crop growth stage,
   - weather conditions.
3. Remove irrelevant content.
4. Do NOT rewrite or interpret management steps. Only include raw context fragments.
5. Do NOT add any new information.

Return the extracted information in this structure:

**Plant:** {plant}
**Disease:** {disease}
**Crop Stage Provided:** {crop_stage}
**Weather Conditions Provided:** {weather}

**Symptoms (from context):**
- Bullet points of symptom text found in context.

**Cause / Pathogen (from context):**
- Extract exact lines mentioning pathogen or cause.

**Spread / Favorable Conditions (from context):**
- Include only lines related to environment, humidity, rain, wind, temperature, leaf wetness, etc.

**Crop StageSpecific Information (from context):**
- Extract any text related to crop stage, growth phase, phenology, etc.

**Weather-Dependent Information (from context):**
- Extract any text related to how weather influences development or severity.

**Management Info Found (raw context only):**
- List raw context lines about management, without rewriting or interpreting them.

If no relevant information is found for any section, write:
"Not available in retrieved context."

If nothing matches at all, write:
"No relevant information found in retrieved context."

If something out of the CONTEXT just say "there is no mentioning about the plant disease."

<|eot_id|><|start_header_id|>user<|end_header_id|>
PLANT: {plant}
DISEASE: {disease}
CROP STAGE: {crop_stage}
WEATHER: {weather}

CONTEXT:
{context}

Extract only the relevant pieces and return the structured summary above.

<|eot_id|><|start_header_id|>assistant<|end_header_id|>
""",
    input_variables=["plant", "disease", "crop_stage", "weather", "context"],
)
```

```python
rag_prompt_chain = rag_context_agent_prompt | GROQ_LLM | StrOutputParser()

QUESTION = """Give me everthing about Tomato yellow leaf curl virus"""
CONTEXT = retriever.invoke(QUESTION)

result = rag_prompt_chain.invoke({"plant": "Tomato", "disease":"yellow leaf curl virus", "crop_stage":"fruit development", "weather": "High humidity, cloudy weather, frequent rainfall", "context":CONTEXT})

print(result)
```
