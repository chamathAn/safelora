```python
import tensorflow as tf
from tensorflow import keras

import matplotlib.pyplot as plt
import numpy as np

import os
```

## Globals


```python
image_size = 224
target_size = (image_size, image_size)
input_shape = (image_size, image_size, 3)

batch_size = 32
epochs = 25
```

## Path Configurations

```python
base_dir = "Datasets/PlantVillage(Arguemnted)"
train_dir = os.path.join(base_dir,"train")
test_dir = os.path.join(base_dir,"test")
```

## Preprocessing

```python
train_datagen = keras.preprocessing.image.ImageDataGenerator(rescale = 1/255.0,         # normalizing
                                                             shear_range = 0.2,         # randomly adding shear transformation
                                                             zoom_range = 0.2,          # randomly adding zooming inside pictures
                                                             width_shift_range = 0.2,   # range of width of the randomly translated pictures
                                                             height_shift_range = 0.2,  # range of height of the randomly translated pictures
                                                             fill_mode="nearest"        # strategy to filling newly adding pixels because of image transformation(change of width and height)
                                                             )

test_datagen = keras.preprocessing.image.ImageDataGenerator(rescale = 1/255.0)
```

## Prepare Data

`train_datagen.flow_from_directory()` class will automatically create fearures and labels as batches that are corresponding to the folder name.


```python
train_data = train_datagen.flow_from_directory(train_dir,
                                               target_size = (image_size, image_size),
                                               batch_size = batch_size,                 # batch size
                                               class_mode = "categorical")              # mode

test_data = test_datagen.flow_from_directory(test_dir,
                                             target_size = (image_size, image_size),
                                             batch_size = batch_size,
                                             class_mode = "categorical")
```

## Create JSON file to store classes with coresponding index

```python
categories = list(train_data.class_indices.keys())
print(train_data.class_indices)
```

```python
import json
with open('class_indices.json','w') as f:
  json.dump(train_data.class_indices, f)

from IPython.display import FileLink
FileLink(r'class_indices.json')
```

## Training

### Load the mobilenet architecture

load with pretrained weights of imagenet

```python
base_model = tf.keras.applications.MobileNet(weights = "imagenet",
                                             include_top = False,
                                             input_shape = input_shape)

base_model.trainable = False
```

Adding Final layer to base MobileNet Model

```python
inputs = keras.Input(shape = input_shape)

x = base_model(inputs, training = False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(len(categories),
                          activation="softmax")(x)

model = keras.Model(inputs = inputs,
                    outputs = x,
                    name="PlantVillage_MobileNet_Model")
```

compile the model

```python
optimizer = tf.keras.optimizers.Adam()

model.compile(optimizer = optimizer,
              loss = tf.keras.losses.CategoricalCrossentropy(from_logits = True),
              metrics=[keras.metrics.CategoricalAccuracy(),
                       'accuracy'])
```

Train

```python
history = model.fit(train_data,
                    validation_data=test_data,
                    epochs=epochs,
                    steps_per_epoch=150,
                    validation_steps=100)
```

## Review the training process

```python
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

## Save the model

```python
model.save('PlantVillage_MobileNet_Model.keras')
```
